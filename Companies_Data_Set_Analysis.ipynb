**1. Data Preprocessing**

```python
#Check unique value 
clients.select_dtypes(include='object').nunique()
#select_dtype = untuk memilih kolom-kolom dalam dataframe yang bertipe data tertentu
#include='object', parameter untuk memilih kolom bertipe data objek (string)
```

```python
clients.dtypes
#Mengecek tipe data setiap kolom
```

```python
#Merubah tipe data yang belum sesuai
# Consulting sales, software sales, forecast growth, returns --> int
# Month, day, year, location, account --> kategori
# Hapus terlebih dahulu karakter "IDR" sebelum mengkonvert object ke int
clients[['Consulting Sales', 'Software Sales','Returns']]=\
clients[['Consulting Sales', 'Software Sales','Returns']].apply(lambda x: x.str.replace('IDR',''))
#Lambda untuk membuat fungsi sederhana yang ditujukkan untuk sebuah elemen saja

#Hapus karakter ',' pada kolom returns
clients['Returns']=clients['Returns'].str.replace(',','')
```

```python
#Menghilangkan karakter '%'
clients['Forecasted Growth']=clients['Forecasted Growth'].str.replace('%','')
```

```python
clients['Forecasted Growth']= clients['Forecasted Growth'].astype('float') * 100 
#Merubah tipe data menjadi float lalu dikali 100 untuk mendapat nilai persenannya

clients['Forecasted Growth']= clients['Forecasted Growth'].astype('int64')
#Merubah tipe data menjadi int
```

```python
clients.dtypes
```

```python
#Ubah tipe data menjadi int

clients[['Consulting Sales', 'Software Sales','Returns']]=\
clients[['Consulting Sales', 'Software Sales','Returns']].astype('int64')
```

```python
#Ubah tipe data menjadi kategori

clients[['Month', 'Day', 'Year', 'Location', 'Account']]=\
clients[['Month', 'Day', 'Year', 'Location', 'Account']].astype('category')
```

```python
#Cek kembali tipe data 

clients.dtypes
```

#Case 1
Sebagai seorang Data Analyst, Anda diminta untuk mendapatkan total penjualan secara keseluruhan dengan mengakumulasikan nilai Consulting Sales dan Software Sales dari setiap perusahaan. Buatlah kolom baru bernama Total Sales yang menyimpan total dari kedua nilai tersebut.
